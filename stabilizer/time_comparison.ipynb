```python
import matplotlib.pyplot as plt
import numpy as np

def read_simulation_data(filename):
    rounds = []
    qubits = []
    method1 = []
    method2 = []

    with open(filename, 'r') as file:
        next(file) 
        for line in file:
            data = line.split()
            rounds.append(int(data[0]))
            qubits.append(int(data[1]))       
            method1.append(float(data[2]))   
            method2.append(float(data[3]))   

    return np.array(rounds), np.array(qubits), np.array(method1), np.array(method2)
```

```python

rounds, qubits, NWQ, Clifford = read_simulation_data("qubits_time_HCNOT.txt")

plt.figure(figsize=(8, 5))
plt.plot(qubits, NWQ, label="SV Sim", marker='o')
plt.plot(qubits, Clifford, label="STAB Sim", marker='s')

plt.title("Statevector vs Stabilizer Simulation of Shallow Circuits")
plt.xlabel("# Qubits/Circuit Width")
plt.ylabel("Time (seconds)")
plt.legend()

plt.show()
```

```python
rounds, qubits, NWQ, Clifford = read_simulation_data("qubits_time_rounds.txt")

plt.figure(figsize=(8, 5))
plt.plot(rounds, NWQ, label="SV Sim", marker='o')
plt.plot(rounds, Clifford, label="STAB Sim", marker='s')

plt.title("Statevector vs Stabilizer Deep Circuits, 10 Qubits Wide")
plt.xlabel("# of 'Extraction' Rounds (Repeated Clifford Operations)")
plt.ylabel("Time (seconds)")
plt.legend()

plt.show()
```
